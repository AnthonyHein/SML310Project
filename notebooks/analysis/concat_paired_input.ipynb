## Setup

```python
from datetime import datetime
import git
import os
import re
from typing import List
from sklearn.preprocessing import StandardScaler
from tqdm import tqdm
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_theme(style="whitegrid")
```

```python
BASE_DIR = git.Repo(os.getcwd(), search_parent_directories=True).working_dir
BASE_DIR
```

---

```python
NUM_PARTITIONS = 50
```

```python
df = pd.read_csv(f"{BASE_DIR}/data/streamline/paired/horses_featurized_jocket_paired_input_0.csv",
                 low_memory=False) 

for i in tqdm(range(1, NUM_PARTITIONS)):
    
    df = pd.concat([
        df,
        pd.read_csv(f"{BASE_DIR}/data/streamline/paired/horses_featurized_jocket_paired_input_{i}.csv",
                    low_memory=False)
    ])
```

```python
df.to_csv(f"{BASE_DIR}/data/streamline/horses_featurized_jockey_paired_input.csv", index=False)
```
